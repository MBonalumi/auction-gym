```python
import numpy as np
from utils import get_project_root as ROOT
import joblib
```

```python
clairevoyant_filename = "contextual_bestbid_5arms.joblib"
clairevoyant_path = ROOT() / "src" / "clairevoyants" / clairevoyant_filename
```

```python
clairevoyant_model = joblib.load(clairevoyant_path)
```

```python
rng = np.random.default_rng(0)
contexts = rng.normal(0.,1., (100,5))
contexts = np.array([np.concatenate((c, [1.0])) for c in contexts])
print(contexts.shape)
```

```python
discrete_space = np.array([-1.09, 0.0, 1.09])  # centroids of a gaussian divided in 3, found through analysis in `Testing Stuff.ipynb`
bin_separator = np.array([-0.4307, 0.4307])

discretized_contexts = []

for context in contexts:
    discr_context = discrete_space[np.digitize(context[:-1], bins=bin_separator)]
    discr_context = np.concatenate((discr_context, [1.0]))
    print(f"Context: {context}, true context: {discr_context}")

    discretized_contexts.append(discr_context)

discretized_contexts = np.array(discretized_contexts)
```

```python
print(discretized_contexts[0])
contexts_samples = rng.choice(discretized_contexts, 10)
print(contexts_samples)
```

```python
print(clairevoyant_model.predict(contexts_samples))
# print(contexts[0][0], clairevoyant_model.predict(contexts[0][0].reshape(-1,1)))
```
